```python
# 1. Load libraries & Dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

# For inline plots (if using Jupyter)
%matplotlib inline
```

```python
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head(20)
```

```python
# 2. Basic Information & Overview
print("Shape of the dataset:", df.shape)
print("\nData Info:\n")
print(df.info())

print("\nSummary Statistics (Numerical):")
display(df.describe())

print("\nSummary Statistics (Including Categorical):")
display(df.describe(include='all'))
```

### There are 150 rows with 5 columns in the Iris dataset

- 4 numerical feature columns (measurements of flowers)
- 1 categorical target column (`species`)

```python
# 3. Unique values and missing values for each column
for col in df.columns:
    print(f"\nColumn: {col}")
    print("Unique values:", df[col].unique()[:10], "...")  # show first 10 unique values
    print("Number of unique values:", df[col].nunique())
    print("Missing values:", df[col].isna().sum())
```

### Columns and their data values

- **sepal length (cm)** → Numerical (Continuous)
- **sepal width (cm)** → Numerical (Continuous)
- **petal length (cm)** → Numerical (Continuous)
- **petal width (cm)** → Numerical (Continuous)
- **species** → Categorical (Nominal)

```python
# 4. Missing values overview
print(df.isnull().sum())
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# 5. Univariate Analysis - Numerical Features

num_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# 6. Skewness of numerical features
for col in num_cols:
    print(f"{col} skew: {df[col].skew():.3f}")
```

```python
# 7. Univariate Analysis - Categorical Feature (species)
sns.countplot(x='species', data=df)
plt.title("Count of each species")
plt.show()
```

```python
# 8. Encoding Categorical Variable

from sklearn.preprocessing import LabelEncoder

# Label Encoding for species
le = LabelEncoder()
df['species_label'] = le.fit_transform(df['species'])
print("Species classes:", le.classes_)
df[['species', 'species_label']].head()
```

```python
# One-Hot Encoding for species
df_ohe = pd.get_dummies(df['species'], prefix='species', drop_first=True)
df_ohe.head()
```

```python
# 9. Bivariate Analysis - Numerical vs Categorical (species)

for col in num_cols:
    plt.figure()
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"{col} by species")
    plt.show()
```

```python
# 10. Multivariate Analysis - Pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
# 11. Correlation and Multivariate Analysis

corr_matrix = df[num_cols.tolist() + ['species_label']].corr()
print(corr_matrix)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 12. Outlier Detection using IQR for numerical features

def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers

for col in num_cols:
    outliers = detect_outliers_iqr(df[col])
    print(f"\nColumn: {col}")
    print("Number of outliers:", len(outliers))
```

```python
# 13. Feature Engineering

# Create ratio features
df['sepal_ratio'] = df['sepal length (cm)'] / df['sepal width (cm)']
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']

# Create petal size level (similar to 'Age_Group' / 'Income_Level' idea)
petal_length_bins = [0, 2.5, 4.5, 7]
petal_length_labels = ['Short', 'Medium', 'Long']
df['Petal_Length_Level'] = pd.cut(df['petal length (cm)'], bins=petal_length_bins, labels=petal_length_labels)

# Visualize Petal Length Level by species
sns.countplot(x='Petal_Length_Level', hue='species', data=df)
plt.title("Petal Length Level by Species")
plt.show()

df[['petal length (cm)', 'Petal_Length_Level', 'species']].head()
```

### Insights:

- Iris-setosa generally has smaller petal length and width compared to the other species.
- Iris-virginica often shows higher values for petal length and width.
- Strong correlations exist between petal length, petal width, and the target (`species_label`).
- The engineered ratio features and petal length levels may help some models distinguish between species more clearly.
