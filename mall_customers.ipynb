## 5.Data Cleaning


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Mall_Customers.csv')  # Adjust path as needed

# 1. Check for missing values
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f'Duplicates: {duplicates}')
# If duplicates exist, remove them
df = df.drop_duplicates()

# 4. Verify data types
print(df.dtypes)

df.rename(columns={
    'Annual Income (k$)': 'Annual_Income',
    'Spending Score (1-100)': 'Spending_Score'
}, inplace=True)


df = df[(df['Age'] >= 18) & (df['Age'] <= 70)]

# Final shape and info
print(df.shape)
print(df.info())

```

#### Insights

No missing values or duplicates in the data.

Dataset has 200 customers with 5 columns.

Four columns are numeric, one is categorical (Genre).

Age, Income, and Spending Score have clean, valid values.

Dataset is small and easy to work with for analysis

## 6.Analysis with Numpy and Pandas

```python
import numpy as np
import pandas as pd

# Assuming df is already loaded and cleaned, with renamed columns

# Average Age of Customers
avg_age = np.mean(df['Age'])
print("Average Age:", round(avg_age, 2))

# Average Annual Income
avg_income = np.mean(df['Annual_Income'])
print("Average Annual Income (k$):", round(avg_income, 2))

# Average Spending Score
avg_spending = np.mean(df['Spending_Score'])
print("Average Spending Score:", round(avg_spending, 2))

# Average Spending Score by Genre
spending_by_genre = df.groupby('Genre')['Spending_Score'].mean()
print("\nAverage Spending Score by Genre:")
print(spending_by_genre)

# Average Annual Income by Genre
income_by_genre = df.groupby('Genre')['Annual_Income'].mean()
print("\nAverage Annual Income (k$) by Genre:")
print(income_by_genre)
```

#### Insights
The average customer is about 39 years old.

Average annual income is around $60.56k.

The average spending score is 50.2, showing moderate spending habits.

Female customers tend to have a slightly higher average spending score than males.

Male customers have a slightly higher average annual income than females.

## 7.Visualization with Matplotlib

```python
import matplotlib.pyplot as plt

# Data
genres = ['Female', 'Male']
avg_spending = [51.53, 48.51]
avg_income = [59.25, 62.23]

plt.figure(figsize=(12, 5))

# Bar Chart: Average Spending Score by Genre
plt.subplot(1, 2, 1)
plt.bar(genres, avg_spending, color=['pink', 'blue'])
plt.title('Average Spending Score by Genre')
plt.xlabel('Genre')
plt.ylabel('Spending Score')
plt.ylim(0, 100)

# Line Plot: Average Annual Income by Genre
plt.subplot(1, 2, 2)
plt.plot(genres, avg_income, marker='o', color='green', linestyle='-', linewidth=2)
plt.title('Average Annual Income (k$) by Genre')
plt.xlabel('Genre')
plt.ylabel('Annual Income (k$)')
plt.ylim(0, 100)

plt.tight_layout()
plt.show()
```

#### Insights:
Females spend more on average than males.

Males have higher average annual income.

Spending habits differ by gender despite income differences.

Both genders show moderate spending scores overall.

## 8.Visualization with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn style
sns.set(style="whitegrid")

# 1. Histogram: Distribution of Annual Income
plt.figure(figsize=(8, 5))
sns.histplot(df['Annual_Income'], bins=15, kde=True, color='teal')
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Count')
plt.show()

# 2. Heatmap: Correlation matrix
plt.figure(figsize=(8, 5))
corr = df[['Age', 'Annual_Income', 'Spending_Score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 3. Boxplot: Spending Score by Genre
plt.figure(figsize=(8, 5))
sns.boxplot(x='Genre', y='Spending_Score', data=df, palette='pastel')
plt.title('Spending Score by Gender')
plt.show()

# 4. Pairplot: Relationships between Age, Income, and Spending
sns.pairplot(df[['Age', 'Annual_Income', 'Spending_Score']], diag_kind='kde')
plt.suptitle('Pairplot of Key Numerical Features', y=1.02)
plt.show()
```

#### Insights:
1. Histogram – Annual Income Distribution

Most customers earn between $40k and $80k per year.

The distribution is fairly uniform, with no strong skew.

A few customers earn very low (<$30k) or very high (>$100k) incomes, indicating a wide income range.

 2. Heatmap – Correlation Matrix

Spending Score and Annual Income have no strong correlation (near 0).

Age and Spending Score show a slightly negative correlation, suggesting older customers may spend less.

Age and Income have a weak positive correlation, meaning older customers tend to earn a bit more.

3. Boxplot – Spending Score by Gender

Female customers have a higher median spending score than males.

The spread (IQR) for both genders is wide, showing diverse spending behavior.

Some outliers exist, especially for high spenders in both groups.

 4. Pairplot – Age, Income, and Spending Score

No strong linear relationship between any pair of features.

Younger customers show a wider range of spending scores, from low to very high.

Customers with moderate income (40k–80k) are scattered across all spending score levels.
